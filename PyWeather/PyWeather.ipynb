```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
import json
```

```python
# Import the census data into a pandas DataFrame
cities_pd = pd.read_csv("worldcities.csv")

# Preview the data
```

```python
# ANOTHER ALTERNATE way TO SELECT THE CITY
#cities=cities_pd
#cities = cities[ (cities["Latitude"].astype(int) > -10.0) & (cities["Latitude"].astype(int) < 10.0)]
#cities.head()
```

```python
selected_cities=cities_pd
selected_cities["selected"]="no"
lat_north = 40.0
lat_south = -40.0


for index, row in selected_cities.iterrows():
    try:
        #print(row["City"] +" = " + str(row["Latitude"]))
        #if (int(row["Latitude"]) > -40.0 and int(row["Latitude"]) < 40.0):
        
        if (float(row["Latitude"]) > lat_south and float(row["Latitude"]) < lat_north):
            selected_cities.set_value(index, "selected", "yes")
            print("Set Value="+ row["City"] +" " + str(row["Latitude"]) + " " + row["selected"] )
    except:
        print("error")
        continue
    


selected_cities.reset_index(inplace=True)
selected_cities.head()
#for index, row in selected_cities.iterrows():
#    if row["selected"] == "yes":
#        print(str(index) + " " + row["City"] + " " + str(row["Latitude"]))
```

```python
# Randomly select 500 zip codes locations that have at least 100 residents
random_selected_cities = selected_cities.sample(n=1000)
random_selected_cities = random_selected_cities[ (random_selected_cities["selected"] == "yes")]

random_selected_cities.head()
```

```python
########################################################################
###
###  Getting DataSet ready - Temperature. Unit Default: Kelvin, Metric: Celsius, Imperial: Fahrenheit.
###
#######################################################################
url = "http://api.openweathermap.org/data/2.5/weather?"

apiKey = "41d4d1d84c8be90f46356a874ecdf66e"

print ("Total number of sampling cities : " + str(len(random_selected_cities)) + " for latitude between " \
       + str(lat_south) + " and " + str(lat_north) )

Latitude_x=[]
Humidity_y=[]
Temperature_y=[]
Cloudiness_y=[]
WindSpeed_y = []
units = "imperial"
recordNumber = 0
setNumber = 1
output_file=[]
output_file_columns=["city","latitude","temperature (F)","humidity (%)","cloudiness (%)","wind speed (mph)"]
current_date = time.strftime("%m/%d/%Y")

for index, row in random_selected_cities.iterrows():
    if recordNumber < 50 :
        recordNumber += 1
    else:
        recordNumber = 0
        setNumber += 1
      
    url = url + "appid=" + apiKey + "&q=" + str(row[2]).replace(" ","%20") + "&units=" + str(units)
    print ("\nCity : " + str(row[2]) + "\n" + "Processing Record " + str(recordNumber) + " of Set " + str(setNumber) + "\n" + url)
    response =  requests.get(url)
    response_json = response.json()
    #print(json.dumps(response_json, indent=4, sort_keys=True))
    try:
        if response_json["cod"] != "404":
            Latitude_x.append((response_json["coord"]["lat"]))
            Temperature_y.append((response_json["main"]["temp_max"]))
            Humidity_y.append((response_json["main"]["humidity"]))
            Cloudiness_y.append((response_json["clouds"]["all"]))
            WindSpeed_y.append((response_json["wind"]["speed"]))
            if response_json["coord"]["lat"] > 40.0 :
                print(row[2])
            output_file.append({"city":str(row[2]),
                                "latitude"    : response_json["coord"]["lat"],
                                "temperature (F)" : response_json["main"]["temp_max"],
                                "humidity (%)"    : response_json["main"]["humidity"],
                                "cloudiness (%)" : response_json["clouds"]["all"],
                                "wind speed (mph)"  : response_json["wind"]["speed"]
                               })

            
            #print("latitude= " + str(response_json["coord"]["lat"]))
            #print("Humidity= " + str(response_json["main"]["humidity"]))
            
        
        #else:
         #   print(str(row[2]).replace(" ","%20") + " Not Found.")  
        url = "http://api.openweathermap.org/data/2.5/weather?"
    except:
        print ("Exception Occurs for " + str(row[2]).replace(" ","%20"))
        url = "http://api.openweathermap.org/data/2.5/weather?"
        continue
```

```python
########################################################################
###
###  Temperature (F) vs. Latitude
###
########################################################################


               
plt.title("Temperature vs. City Latitude" + " - " + current_date)
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.scatter( Latitude_x , Temperature_y, s=20, alpha=0.75, linewidths=1.0, edgecolors="blue",facecolors="lightcoral")
plt.savefig("temperature.png")
plt.show()
```

```python
########################################################################
###
###  Humidity (%) vs. Latitude
###
########################################################################
plt.title("Humidity (%) vs. City Latitude"  + " - " + current_date)
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.scatter( Latitude_x , Humidity_y, s=20, alpha=0.75, linewidths=1.0, edgecolors="blue",facecolors="gold")
plt.savefig("humidity.png")
plt.show()
```

```python
########################################################################
###
###  Cloudiness (%) vs. Latitude
###
########################################################################

plt.title("Cloudiness (%) vs. City Latitude" + " - " + current_date)
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.scatter( Latitude_x , Cloudiness_y, s=20, alpha=0.75, linewidths=1.0, edgecolors="blue",facecolors="lightcoral")
plt.savefig("cloudiness.png")
plt.show()
```

```python
########################################################################
###
###  Wind Speed (mph) vs. Latitude
###
########################################################################

plt.title("Wind Speed (mph) vs. City Latitude" + " - " + current_date)
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.scatter( Latitude_x , WindSpeed_y, s=20, alpha=0.75, linewidths=1.0, edgecolors="blue")
plt.savefig("wind_speed.png")
plt.show()
```

```python
csv_df = pd.DataFrame(output_file, columns=output_file_columns)

csv_df.set_index("city", inplace=True)

csv_df.to_csv("PyWeather_output.csv")
csv_df.head()
```
